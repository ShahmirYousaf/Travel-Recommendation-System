```python
import pandas as pd
import os

folder_path = r'D:\Datasets'

# Specify the file name
final_file = 'yelp_dataset_dataset_business_clean_withoriginalfeature.csv'

final_path = os.path.join(folder_path, final_file)
data = pd.read_csv(final_path)

def recommendations(data, state, city, category, day, kids=False):
    recommendation = data[(data['state'] == state) & 
                          (data['city'] == city) & 
                          (data[category] == True) & 
                          (data[day] == True) & 
                          (data['review_count'] > 100) & 
                          (data['stars'] >= 4.5)].sort_values(by='review_count', ascending=False)
    
    if kids == 'Yes':
        recommendation = recommendation[recommendation['GoodForKids'] == True]
    
    return recommendation[['name', 'stars', 'review_count', 'address']]

def new_user_recommendation(data):
    state = input(f'Please enter state:')
    city = input(f'Please enter city:')
    print("Top categories: 'Restaurants','Food','Shopping','Beauty & Spas','Home Services','Health & Medical','Automotive','Nightlife','Local Services','Bars','Event Planning & Services','American (Traditional)','Active Life'")
    category = input("Tell me the category you're interested in:")
    day = input('Please enter day:')
    kids = input('Will you bring kids with you? (Yes/No):')
    
    recommendations_df = recommendations(data, state, city, category, day, kids)
    if not recommendations_df.empty:
        print(recommendations_df[['name', 'stars', 'review_count', 'address']])
    else:
        print("Sorry, no recommendations found for your criteria.")

new_user_recommendation(data)
```
